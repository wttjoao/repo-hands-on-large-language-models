How we can generate contextualized word embeddings

```python
from transformers import AutoModel, AutoTokenizer
```

```python
# Load a tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")

# Load a language model
model = AutoModel.from_pretrained("microsoft/deberta-v3-base")

# Tokenize the sentence
tokens = tokenizer("Hello world", return_tensors="pt")

# Process the tokens
output = model(**tokens)[0]
```

This code downloads a pretrained tokenizer and model, then uses them to process the string "Hello world".

The output of the model is then saved in the output variable.

```python
output.shape
```

- Four tokens, each one embedded in a vector of 768 values
- The first dimension is the batch dimension used in cases when we want to send multiple input sentences to the model at the same time.

What are those 4 tokens?

```python
for token in tokens['input_ids'][0]:
    print(tokenizer.decode(token))
```
